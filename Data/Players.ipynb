```python
#Domed Stadiums
Dome_Dict = {
    "ARI" : 1,
    "ATL" : 1,
    "DAL" : 1,
    "DET" : 1,
    "HOU" : 1,
    "IND" : 1,
    "LAC" : 1,
    "LAR" : 1,
    "LVR" : 1,
    "MIN" : 1,
    "NOR" : 1,
}
```

```python
#Dome Game
df['Dome_Home'] = df['Team'].apply(lambda x: 1 if x in Dome_Dict.keys() else 0)
df['Dome_Away'] = df['Opp'].apply(lambda x: 1 if x in Dome_Dict.keys() else 0)
```

```python
df.head()
```

```python
df['Dome'] = df['Dome_Home'] + df['Dome_Away']
```
